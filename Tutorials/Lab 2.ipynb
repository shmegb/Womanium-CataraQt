# Applying Hadamrd, Two Qubits and Applying other Gates

In this session we'll implement different gates. In the last lab we have imlemented X gate. We will be looking at Two qubits and  operations on them.

```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
q =  QuantumRegister(1)

# Here we are using a single Qubit.
# Here we need a Classical register to store our information.
# we are using single CBit.

c = ClassicalRegister(1)


# Quantum circuit is composed of classical and quantum register.
# Let's define our Quantum circuit

qc = QuantumCircuit(q,c)

# Now we are applying gates on our qubit, NOT operator/ gate.
qc.h(q[0])
# We have applied X (NOT) gate to the first qubit.

# Now we are associating a quantum bit to a classical bit In order to measure our Qubit.
qc.measure(q[0],c[0])
# Whatever we get from a measurement, is now stored in a cbit.


print("The design of the circuit is done.")
```

```python
qc.draw(output='mpl')
```

Hadamard gate creates superposition. 

```python
from qiskit import execute, Aer
# Here we have used method "execute" and "Aer" object from the qiskit library for the execution of our circuit.
# Here we have 3 parameters mycircuit,backend on which it will be executed and how many times it will be executed.

job = execute(qc,Aer.get_backend('qasm_simulator'),shots=1024)

# Result of our outcome will be 
counts = job.result().get_counts(qc)
print(counts) # counts is a dictionary
```

```python
from qiskit.visualization import plot_histogram
plot_histogram(counts)
```

# Two Qubits

In the below example we are using two qubits.

```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit

q =  QuantumRegister(2)

c = ClassicalRegister(2)

circuit = QuantumCircuit(q,c)
# We are applying X gate and then H gate.

# Applying X gate to the first qubit
circuit.x(0)

# Applying H gate to the first qubit
circuit.h(0)

circuit.barrier()

# Measure the circuit
circuit.measure(q,c)

# Draw the circuit
circuit.draw(output='mpl')
```

```python
from qiskit import execute, Aer
# Here we have used method "execute" and "Aer" object from the qiskit library for the execution of our circuit.
# Here we have 3 parameters mycircuit,backend on which it will be executed and how many times it will be executed.

job = execute(circuit,Aer.get_backend('qasm_simulator'),shots=1024)

# Result of our outcome will be 
counts = job.result().get_counts(circuit)
print(counts) # counts is a dictionary
```

Here we get the output state |00> and |01>.

```python
from qiskit.visualization import plot_histogram
plot_histogram(counts)
```

# Other Gates

So far we have seen X and H gates. There are other gates. Pauli Y and Pauli Z.

```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit

q =  QuantumRegister(1)

c = ClassicalRegister(1)

circuit1 = QuantumCircuit(q,c)
# We are applying X gate and then H gate.

# Applying X gate to the first qubit
circuit1.y(0)

# Applying H gate to the first qubit


circuit1.barrier()

# Measure the circuit
circuit1.measure(q,c)

# Draw the circuit
circuit1.draw(output='mpl')
```

```python
from qiskit import execute, Aer
# Here we have used method "execute" and "Aer" object from the qiskit library for the execution of our circuit.
# Here we have 3 parameters mycircuit,backend on which it will be executed and how many times it will be executed.

job = execute(circuit1,Aer.get_backend('qasm_simulator'),shots=1024)

# Result of our outcome will be 
counts = job.result().get_counts(circuit1)
print(counts) # counts is a dictionary
```

```python
from qiskit.visualization import plot_histogram
plot_histogram(counts)
```

Now let see what will happen when we implement z gate.

```python
from qiskit import execute, Aer
# Here we have used method "execute" and "Aer" object from the qiskit library for the execution of our circuit.
# Here we have 3 parameters mycircuit,backend on which it will be executed and how many times it will be executed.

job = execute(circuit2,Aer.get_backend('qasm_simulator'),shots=1024)

# Result of our outcome will be 
counts = job.result().get_counts(circuit2)
print(counts) # counts is a dictionary
```

```python
from qiskit.visualization import plot_histogram
plot_histogram(counts)
```
